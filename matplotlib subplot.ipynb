```python
# subplot is used if we have so many graphs or something extra information then we can use subplot so that it will show in different 

import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x,y,color='r')

plt.subplot(2,2,2)
plt.pie([1],colors='y')


x1=[10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x)

x2=['a','b','c','d']
y2=[10,20,30,40]

plt.subplot(2,2,4)
plt.bar(x2,y2)
plt.show()
```
