# Box Vs Whisker Plot 

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x)

#horizontal is Box plot and vertical is wiskey plot
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)
# notch=True parameter is used in the boxplot function to create a notched box plot. 

#horizontal is Box plot and vertical is wiskey plot
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,vert=False)
#By default it is vertical if we keep vert=False then we get an Horizontal boxplot

#horizontal is Box plot and vertical is wiskey plot
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,tick_labels=['python'],patch_artist=True,showmeans=True)
# tick_labels, which is now the correct way to label the x-axis ticks in a box plot.
# patch_artist means it will give the default color
# To include the mean as a marker on your box plot, you can use the showmeans=True parameter in the plt.boxplot() function. 
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,tick_labels=['python'],showmeans=True,sym='g+')
# sym parameter in plt.boxplot() is used to specify the style of the markers for outliers. 
# circle or + is called as outline

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,tick_labels=['python'],showmeans=True,whis=2)
# whis parameter in plt.boxplot() controls the length of the whiskers in a box plot.
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]


plt.boxplot(x,tick_labels=['python'],showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='blue'),whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y'))
# whiskerprops parameter in plt.boxplot() allows you to customize the appearance of the whiskers in a box plot

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]
x1=[10,20,30,40,50,30,60,90]


plt.boxplot(x,tick_labels=['python'],showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='blue'),whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y'))
# whiskerprops parameter in plt.boxplot() allows you to customize the appearance of the whiskers in a box plot
plt.boxplot(x1,tick_labels=['python'],showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='blue'),whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y'))

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]

y=[x,x1]
plt.boxplot(y,label=['python','c++'],showmeans=True,sym='g+',boxprops=dict(color='r'),capprops=dict(color='blue'),whiskerprops=dict(color='orange'))

plt.show()
```
