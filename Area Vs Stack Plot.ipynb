# Area Vs Stack Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area=[1,2,3,4,5]


plt.stackplot(x,area)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]

l = ['area1', 'area2', 'area3']

# Create the stack plot
plt.stackplot(x, area1, area2, area3, labels=l, colors=['r', 'b', 'g'], baseline='zero')
# in baseline we have sym,wiggle

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.legend(loc=2)  # Location of the label
plt.show()
```
