# Matplotlib Step Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]

y=[1,2,3,4,5]

plt.step(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]

y=[4,2,5,7,0]

plt.step(x,y,marker='o',color='r',ms=10,mfc='g',label='python')
# ms means marker size, and mfc means marker face color


plt.title('python')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.grid()
plt.legend(loc=2)
plt.show()
```
