# Stem Plots with Matplotlib

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=':',markerfmt='r+',bottom=0,basefmt='green',label='python',orientation='horizontal')
# linefmt parameter specifies the format of the stem lines—the vertical lines that extend from the baseline to the data points. 
# markerfmt parameter in the plt.stem() function is used to customize the appearance of the markers at the top of each stem in the plot
# bottom parameter specifies the starting position of the stems along the y-axis.
# basefmt parameter in the plt.stem() function specifies the appearance of the baseline, which is the horizontal line from which the stems originate. By setting basefmt='green', you are specifying that the baseline should be a solid line colored green.


plt.legend()
plt.show()
```
