# Matplotlib Pie Chart

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=['c','c++','Java','python']

ex=[0.3,0.0,0.0,0.0]

c=['red','pink','brown','yellow'] #by default it will give the color but we can change the color of the pie chat

plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow=True,radius=0.9,labeldistance=1.1,startangle=0,textprops={'fontsize': 15},counterclock=False,wedgeprops={'linewidth':5,'edgecolor': 'm'},center=(1,3)) #if we want to write a name for that expendeecture then we can use the label
# explode parameter is used in the pie() function to "pull out" or "explode" a slice of the pie chart away from the center
# The autopct parameter in the pie() function  specifies how the percentage of each slice should be displayed.
# shadow=True parameter in the pie() function of Matplotlib is used to add a shadow effect to the pie chart, giving it a slightly three-dimensional appearance.
# radius parameter in the pie() function of Matplotlib is used to control the size of the pie chart.
# labeldistance parameter in the pie() function of Matplotlib controls the distance between the center of the pie chart and the labels that are placed outside the slices Like c,c++,java,python
# startangle parameter in the pie() function of Matplotlib determines the starting angle of the first pie slice. 
# The textprops parameter in the pie() function of Matplotlib is used to customize the properties of the text labels on the pie chart. You can pass a dictionary to textprops to specify various text properties such as fontsize, color, fontweight, etc.
#  counterclock parameter in the pie() function of Matplotlib controls the direction in which the pie slices are drawn. By default, the slices are drawn counterclockwise.
# wedgeprops parameter in the pie() function of Matplotlib is used to customize the properties of the pie chart slices, or "wedges." You can pass a dictionary to wedgeprops to specify various properties such as linewidth, edgecolor, linestyle, etc.
# center parameter in the pie() function of Matplotlib specifies the position of the center of the pie chart within the plot


#plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=['c','c++','Java','python']

ex=[0.3,0.0,0.0,0.0]

c=['red','pink','brown','yellow'] #by default it will give the color but we can change the color of the pie chat

plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow=True,radius=0.9,labeldistance=1.1,startangle=0,textprops={'fontsize': 15},counterclock=False,wedgeprops={'linewidth':5},rotatelabels=True) #if we want to write a name for that expendeecture then we can use the label
# rotatelabels parameter is not a built-in parameter of Matplotlib's pie() function. However, you might want to rotate the labels in a pie chart to improve readability or to fit them better.


plt.title('Expenditure')
plt.legend(loc=2)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
x1=[40,30,20,10]
y=['c','c++','Java','python']




c=['red','pink','brown','yellow'] #by default it will give the color but we can change the color of the pie chat

plt.pie(x,labels=y,radius=1.5)

cr=plt.Circle(xy=(0,0),radius=1,facecolor='white')
plt.gca().add_artist(cr) #Like this also we can do the ring pie chat...
# plt.Circle to draw a circle, and plt.gca().add_artist() to add this circle to the current plot



plt.show()
```
