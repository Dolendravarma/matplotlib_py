```python
import matplotlib.pyplot as plt

x=[1, 3, 5, 7 ,9]
y=[2, 4, 6 ,8, 10]

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1, 3, 5, 7 ,9]
y=[2, 4, 6 ,8, 10]

plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1, 3, 5, 7 ,9]
y=[2, 4, 6 ,8, 10]

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1, 3, 5, 7 ,9]
y=[2, 4, 6 ,8, 10]

plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1, 3, 5, 7 ,9]
y=[2, 4, 6 ,8, 10]

plt.hist(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1, 2, 8, 7]
y=[2, 4, 6 ,3]

plt.pie(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1, 3, 5, 7 ,9]
y=[2, 4, 6 ,8, 10]

plt.fill_between(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1, 3, 5, 7 ,9]
y=[2, 4, 6 ,8, 10]

plt.step(x,y)
plt.show()
```

```python
#we can add the color aslo to the bars
import matplotlib.pyplot as plt

x=[1, 3, 5, 7 ,9]
y=[2, 4, 6 ,8, 10]

c=['red','green','blue','orange','cyan']


plt.bar(x,y,color=c)
plt.show()
```
