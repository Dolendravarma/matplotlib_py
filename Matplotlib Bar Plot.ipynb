#Bar plot

```python
import matplotlib.pyplot as plt

x=['Python','c','c++','Java']

y=[85, 70, 60 ,82]

plt.xlabel('languages',fontsize=15) #x label is used to set the name of the x axis and font size is used to set the size of this axis 
plt.ylabel('Numbers',fontsize=15) #y label is used to set the name of the y axis
plt.title('prog_lag',fontsize=15) #title is used to set the title

c=['y','blue','magenta','green'] #to get the all colors of bar graph we keep this in the list and wee call them in bar
plt.bar(x,y,width=0.4,color=c,align='center',edgecolor='red',linewidth=5,linestyle=':',alpha=0.5)# width is used to set the bars width here
#align parameter is used to set the content center or edge by default it is in the center
#edgecolor parameter is used to set the color of the bar edge..
#linewidth parameter is used to set the width of the edge...
#linestyle parameter is used to set the style like solid,dotted,etc
#alpha parameter is used to set the light color of the bar inside color it ranges from 0-1 only


plt.show()
```

```python
import matplotlib.pyplot as plt

x=['Python','c','c++','Java']

y=[85, 70, 60 ,82]

plt.xlabel('languages',fontsize=15) #x label is used to set the name of the x axis and font size is used to set the size of this axis 
plt.ylabel('Numbers',fontsize=15) #y label is used to set the name of the y axis
plt.title('prog_lag',fontsize=15) #title is used to set the title

c=['y','blue','magenta','green']

plt.bar(x,y,width=0.4,color=c,align='center',edgecolor='red',label='Prog')
#label parameter is used to set the label like down see the prog like that and we should mention that legend calling

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np #in numpy arrange function will be there so when we have prepared the list then it will be an array

x=['Python','c','c++','Java']

y=[85, 70, 60 ,82]

z=[20 ,30, 40, 50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('languages',fontsize=15) #x label is used to set the name of the x axis and font size is used to set the size of this axis 
plt.ylabel('Numbers',fontsize=15) #y label is used to set the name of the y axis
plt.title('prog_lag',fontsize=15) #title is used to set the title




#insted of x use the p so that 
plt.bar(p,y,width,color='red',align='center',label='popularity')
#label parameter is used to set the label like down see the prog like that and we should mention that legend calling

#To create the multiple bar graphs 
plt.bar(p1,z,width,color='blue',align='center',label='popularity 1')
#when we do this it will come in over lapping but we want it side by side then 


plt.xticks(p+width/2,x,rotation=20) #if we do like this we get in blue color names python,c,c++ so reduse that we divide the width by 2
#rotation is used to rotate the python ,c,c++ x values

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=['Python','c','c++','Java']

y=[85, 70, 60 ,82]

z=[20 ,30, 40, 50]

plt.xlabel('languages',fontsize=15) #x label is used to set the name of the x axis and font size is used to set the size of this axis 
plt.ylabel('Numbers',fontsize=15) #y label is used to set the name of the y axis
plt.title('prog_lag',fontsize=15) #title is used to set the title



plt.barh(x,y,color='red',align='center',label='popularity')
#label parameter is used to set the label like down see the prog like that and we should mention that legend calling

#To create the multiple bar graphs we use the 3 dimensions

#if you want to do in horizontal bar graph keep barh
plt.barh(x,z,color='blue',align='center',label='popularity 1')
#when we do this it will come in over lapping but we want it side by side then 

plt.legend()
plt.show()
```
