#Scatter plot

```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6,7]
nums=[2,3,1,4,5,3,6]
plt.scatter(days,nums,color='red')

plt.title('Scatter plot',fontsize=15)
plt.xlabel('Days',fontsize=15)
plt.ylabel('Nums',fontsize=15)
plt.show()
```

```python
#Taking each point different colors
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6,7]
nums=[2,3,1,4,5,3,6]

c=['red','blue','cyan','green','orange','brown','yellow']
s=[400,200,400,300,200,100,600] # we can change the colors and sizes of different points by defing them like a list 
plt.scatter(days,nums,color=c,sizes=s,alpha=0.8,marker='*',edgecolor='red',linewidth=1)
#alpha is used to decrease the color to light color 
#marker is used to keep it like 
#edgecolor is used

plt.title('Scatter plot',fontsize=15)
plt.xlabel('Days',fontsize=15)
plt.ylabel('Nums',fontsize=15)
plt.show()
```

```python
#Taking each point different colors
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]
nums = [2, 3, 1, 4, 5, 3, 6]

c = [10, 99, 67, 89, 45, 78, 15]  # Values for colormap
s = [400, 200, 400, 300, 200, 100, 600]  # Sizes for each point

plt.scatter(days, nums, c=c, s=s, cmap='viridis', alpha=0.8, marker='o', edgecolor='black')
#so for scatter color and size are not valid parameters so use c,s
#in colormaps so many are there like Accent ,Blues BrBG,BuGn,Grrens,OrRd,Pastel1_r,RdPu,Reds,Dark2_r, etc so so many are there


# Titles and labels
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Nums', fontsize=15)



t=plt.colorbar()  # Add colorbar to show the mapping of c values
t.set_label('Color Intensity',fontsize=15)
#we cant directly access the color bar directly instead of that we want to create a label separetely

plt.show()
```

```python
#Taking each point different colors
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]
nums = [2, 3, 1, 4, 5, 3, 6]
nums2=[3, 2, 4, 5, 8, 9, 7]

c = [10, 99, 67, 89, 45, 78, 15]  # Values for colormap
s = [400, 200, 400, 300, 200, 100, 600]  # Sizes for each point

plt.scatter(days, nums, c=c, s=s, cmap='viridis', alpha=0.8, marker='o', edgecolor='black')
#so for scatter color and size are not valid parameters so use c,s
#in colormaps so many are there like Accent ,Blues BrBG,BuGn,Grrens,OrRd,Pastel1_r,RdPu,Reds,Dark2_r, etc so so many are there

plt.scatter(days, nums2, c=c, s=s,  alpha=0.8, marker='o', edgecolor='black')

# Titles and labels
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Days', fontsize=15)
plt.ylabel('Nums', fontsize=15)



t=plt.colorbar()  # Add colorbar to show the mapping of c values
t.set_label('Color Intensity',fontsize=15)
#we cant directly access the color bar directly instead of that we want to create a label separetely

plt.show()
```
